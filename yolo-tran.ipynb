```python
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

def evaluate_coco(gt_json, dt_json):
    coco_gt = COCO(gt_json)
    coco_dt = coco_gt.loadRes(dt_json)

    coco_eval = COCOeval(coco_gt, coco_dt, "bbox")
    coco_eval.params.iouThrs = [0.5]  
    coco_eval.evaluate()
    coco_eval.accumulate()
    coco_eval.summarize()

# Example usage
ground_truth_json = "D:/BTXRD-Dataset/BTXRD-Yolo/val/labels/coco_annotations.json"
detections_json = "D:/BTXRD-Dataset/BTXRD-Yolo/val/labels/detect.json"
evaluate_coco(ground_truth_json, detections_json)
```
